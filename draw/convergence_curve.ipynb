```python
# draw convergence curves of algs in paper with Pseudocode [1,13,15,20]
problem_set = [1, 13, 15, 20]

data_path = 'D:\\01Code\\ALDes\\draw\\datas\\mats\\ArchSolution\\'
for problem in problem_set:
    mat_path = data_path + f'f{problem}.mat'
    mat_data = scipy.io.loadmat(mat_path)
    mat_data = -mat_data['res'][0]

    plt.figure(figsize=(5, 3))
    g = sns.lineplot(mat_data)
    plt.title(f'Convergence Curve of Alg{problem}')
    plt.xlabel('Iterations')
    plt.ylabel('Performance')

    plt.savefig(f'result_pictures\convergence_curve\Alg{problem}.svg',dpi=300,format="svg",bbox_inches = 'tight')

```
